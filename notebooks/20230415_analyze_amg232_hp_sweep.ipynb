# Check out Results from Smaller Model

```python
import numpy as np
import pandas as pd
import sys
import os
import pickle
```

```python
cv_result_path = "small_model_cv_fit/AMG-232_all/BRD-K64925568-001-01-8_all_2.5.pkl"
```

```python
# load cv results
with open(cv_result_path, 'rb') as f:
    cv_results = pickle.load(f)
```

```python
best_params = []
errors = []
for idx, fold in enumerate(cv_results):
    for param in fold["params"]:
        best_params.append({**param, "fold": idx})
    errors.append({"fold": idx, "test_mse": fold["test_mse"], "val_mse": fold["val_mse"], "train_mse": fold["train_mse"]})
```

```python
best_param_df = pd.DataFrame(best_params)
error_df = pd.DataFrame(errors)
```
