```python
# Whats up with the feature importance scores?
import pandas as pd
import numpy as np
from glob import glob
```

```python
X_test_vals = glob("/Users/nick/Desktop/new_refract/data/test_output/refract_models/amg-232_2.5_model/*X_test.csv")
shap_vals = glob("/Users/nick/Desktop/new_refract/data/test_output/refract_models/amg-232_2.5_model/*shap_values.csv")
```

```python
shap_vals = [pd.read_csv(f) for f in shap_vals]
X_test_vals = [pd.read_csv(f) for f in X_test_vals]
# for shap_vals df, set index as depmap_id
for df in shap_vals:
    df.set_index("depmap_id", inplace=True)
for df in X_test_vals:
    df.set_index("depmap_id", inplace=True)
```

```python
# concat shap_vals
shap_vals = pd.concat(shap_vals)
X_test_vals = pd.concat(X_test_vals)
```

```python
# fill NA with 0
shap_vals = shap_vals.fillna(0)
# fill with median feature value for each feature
X_test_vals = X_test_vals.fillna(X_test_vals.median())
```

```python
# get the mean absolute shap value for each feature
mean_abs_shap = shap_vals.abs().mean()
```

```python
# sort
mean_abs_shap.sort_values(ascending=False)
```
