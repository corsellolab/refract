```python
from refract.trainers import XGBoostTrainer
from refract.feature_selection import get_pearson_correlations
from refract.utils import load_feature_df, load_response_df, intersect_depmap_ids
```

```python
feature_df = load_feature_df(feature_file)
response_df = load_response_df(response_file)
response_df, feature_df = intersect_depmap_ids(response_df, feature_df)
```

```python
trainer = XGBoostTrainer(
        output_dir="./results",
        n_threads=4,
        num_rounds=500,
        early_stopping_rounds=25,
        eta=0.05,  # Custom learning rate
        max_depth=4,  # Custom tree depth
    )
```

```python
results = trainer.train_cross_validation(
    feature_file=feature_file,
    response_file=response_file,
    split_dir=splits,
    n_splits=10
)
```
